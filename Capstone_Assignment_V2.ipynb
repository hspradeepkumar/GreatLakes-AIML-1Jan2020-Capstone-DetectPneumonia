```python
# creating Bounding box on images

# finding number of times image exists in bounding box data file
def get_count(pathname,data_lab):
  count_path=0
  for path in data_lab['PathName']:
    if path==pathname:
      count_path+=1
  return count_path

# function to get image
def get_image(pathname,met_data):
  for i in range(len(met_data)):
    if met_data[i]==pathname:
      img=pydicom.dcmread(met_data[i]) 
      imag=img.pixel_array
  return imag

# function to draw bounding Box on images
def get_bb(imag,data_lab):
  for i in range(len(data_lab)):
    if data_lab.iloc[i]['PathName']==pathname and data_lab.iloc[i]['Target']==1:
      x= int(data_lab.iloc[i]['x'])
      y= int(data_lab.iloc[i]['y'])                                                     
      w= int(data_lab.iloc[i]['width'])
      h= int(data_lab.iloc[i]['height'])
      bb=cv2.rectangle(imag,(x,y),(x+w,y+h),(255,0,0),3)
      plt.imshow(bb,cmap=plt.cm.bone)
    elif data_lab.iloc[i]['PathName']==pathname and data_lab.iloc[i]['Target']==0:
      plt.imshow(imag,cmap=plt.cm.bone)    
```

**Resizing Image**

```python
# function to resize images to build train,test and validation set
Image_width=512
Image_height=512
Num_channels=1

Orig_width=1024
Orig_height=1024

def resize_image(meta_data,data_lab):
  cropped_images=np.zeros((len(meta_data),Image_width,Image_height),dtype=np.float32)  
  for a in range(len(meta_data)):
    #pathname=meta_data[a]
    img=pydicom.dcmread(meta_data[a])
    img_pixel=img.pixel_array
    cropped_img=img_pixel[::2,::2]
    cropped_images[a]=cropped_img 
    for i in range(len(data_lab)):
      if data_lab.iloc[i]['PathName']==meta_data[a] and data_lab.iloc[i]['Target']==1:
        x= int(data_lab.iloc[i]['x'])
        y= int(data_lab.iloc[i]['y'])                                                     
        w= int(data_lab.iloc[i]['width'])
        h= int(data_lab.iloc[i]['height'])
        cropped_images[a]=cv2.rectangle(cropped_images[a],(x,y),(x+w,y+h),(255,0,0),3)
      elif data_lab.iloc[i]['PathName']==meta_data[a] and data_lab.iloc[i]['Target']==0:
        cropped_images[a]=cropped_images[a]
  return cropped_images         
```

```python
def get_target(meta_data,data_lab):
  target_var=np.zeros(len(meta_data))
  for a in range(len(meta_data)):
    #pathname=meta_data[a]
    for i in range(len(data_lab)):
      if data_lab.iloc[i]['PathName']==meta_data[a] and data_lab.iloc[i]['Target']==1:
        target_var[a]=data_lab.iloc[i]['Target']
      elif data_lab.iloc[i]['PathName']==meta_data[a] and data_lab.iloc[i]['Target']==0:
        target_var[a]=data_lab.iloc[i]['Target']
  return target_var
```

```python
# function to normalize the bounding box and create dataframe for training, test and validation set

def norm_bb_dataset(meta_d):
  data_labels_norm=pd.DataFrame(columns=data_labels1.columns)
  for a in range(len(meta_d)):
    for i in range(len(data_labels1)):
      if meta_d[a]==data_labels1.iloc[i]['PathName'] and data_labels1.iloc[i]['Target']==1:
        data_labels_norm=data_labels_norm.append({'PathName':data_labels1.iloc[i]['PathName'],'x': int(data_labels1.iloc[i]['x']*Image_width/Orig_width),'y': int(data_labels1.iloc[i]['y']*Image_height/Orig_height),'width':int(data_labels1.iloc[i]['width']*Image_width/Orig_width),'height': int(data_labels1.iloc[i]['height']*Image_height/Orig_height),'Target':int(data_labels1.iloc[i]['Target'])},ignore_index=True)
      elif meta_d[a]==data_labels1.iloc[i]['PathName'] and data_labels1.iloc[i]['Target']==0:
        data_labels_norm=data_labels_norm.append({'PathName':data_labels1.iloc[i]['PathName'],'x': data_labels1.iloc[i]['x'],'y': data_labels1.iloc[i]['y'],'width':data_labels1.iloc[i]['width'],'height': data_labels1.iloc[i]['height'],'Target':data_labels1.iloc[i]['Target']},ignore_index=True)
  return data_labels_norm
```

```python
def save_labels():
    labels = pd.read_csv('stage_2_train_labels.csv')
    #patset=[]
    patbbsetfin=[]
    patient_id=labels['patientId'].value_counts().index
    for pid in patient_id:
         patdtl=labels[labels['patientId']==pid]
         patbbset=[]
         for patrow in patdtl.iterrows():
             bbset=[]
             #print(patrow[1])
             bbset.append(patrow[1][1])
             bbset.append(patrow[1][2])
             bbset.append(patrow[1][3])
             bbset.append(patrow[1][4])
             patbbset.append(bbset) 
         patbbsetfin.append([pid, patbbset, patrow[1]['Target']])

    patbbsetfinarr=np.array(patbbsetfin)
    np.save('patient_labels.npy', patbbsetfinarr) 
```
